###### Testing the Stationary 

Ideally we want to have a stationary timeseries, therefore we will use the Augmented Dicky Fuller test to test our following hypothesis:
- H0  the time series is not stationary
- H1  time series is stationary

```python
from statsmodels.tsa.stattools import adfuller
#setting plot size
fig = plt.figure(figsize=(15, 5))

#defining function for conducting the ADF Test
def adf_test(x):
    
    #rolling statistics
    rolmean = x.rolling(window = 22, center = False).mean() #rolling mean
    rolstd  = x.rolling(window = 12, center = False).std() #rolling standard deviation
    
    #plot rolling statistics
    series = plt.plot(x,       color = 'black',  label = 'Close Price') 
    mean   = plt.plot(rolmean, color = 'red',   label = 'Rolling Mean')
    std    = plt.plot(rolstd,  color = 'blue', label = 'Rolling Std')
    plt.legend()
    plt.title('Bitcoin Close Price Rolling Mean and Standard Deviation')
    plt.show()
    
    #performing ADF Test
    result = adfuller(x)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    for key, value in result[4].items():
        if result [0] > value:
            print('Fail to Reject Null Hypothesis')
        else:
            print('Accept Alternative Hypothesis')
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t', key, ':', value)
        
# Perform the ADF test on the Bitcoin close price series

adf_test(close_df['Close'])
```

The p-value is greater than 0.05, thus Fail to reject the null hypothesi H0 time series is not stationary. 

##### Stationary Transformations

The data is highly skewwed, log transportation is perform to help the forcasting data to achive stationary time series

```python
fig = plt.figure(figsize=(15, 5))

#Log Transformation
log_ts = np.log(close_df['Close'])
plt.plot(log_ts, label = 'Close Price')
plt.title('Log Transform Bitcoin Closing Price')
plt.legend()
plt.show()

fig = plt.figure(figsize=(15, 5))

adf_test(log_ts)
```

The p value is greater than 0.5, we will continue to remove trend and seasonality

```python
#Differencing 
fig = plt.figure(figsize=(15, 5))
diff_log_ts = log_ts - log_ts.shift()
plt.plot(diff_log_ts)
plt.show()

diff_log_ts.dropna(inplace=True)

fig = plt.figure(figsize=(15, 5))

adf_test(diff_log_ts)
```

After various transformation, now that the p-value is below 0.05,the time series is stationary,

```python
import statsmodels.api as sm


# Calculate the autocovariance
autocovariance = sm.tsa.acovf(diff_log_ts)

# Plot the autocovariance
plt.plot(autocovariance)
plt.xlabel('Lag')
plt.ylabel('Autocovariance')
plt.title('Autocovariance Plot')
plt.show()
```

The plot exhbitis Autocovariance deca, meaning where Autocovariance value decrease as the lag increase and the strenght  relationship between observations diminish quickly. This behavior provides further evidence that the time series is non-stationary.

A negative autocovariance for Bitcoin prices suggests a negative linear relationship between the current price and its lagged values. However, it's important to note that interpreting autocovariance in the context of Bitcoin prices can be challenging due to the unique characteristics and dynamics of cryptocurrency markets. Here are a few potential interpretations:

Mean Reversion: Negative autocovariance in Bitcoin prices could indicate mean-reverting behavior. In this case, when the Bitcoin price is above its average, it tends to decrease in subsequent periods, and vice versa. This could imply that extreme price movements are likely to be followed by corrections toward the long-term average price.

Selling Pressure: Negative autocovariance might suggest that periods of high Bitcoin prices are followed by periods of selling pressure or profit-taking. When prices increase, some market participants may choose to sell their Bitcoin holdings, causing the price to decrease in the following periods.

Negative Trend: Negative autocovariance could indicate a declining trend in Bitcoin prices over time. If the autocovariance consistently shows negative values at various lags, it might suggest a downward movement or a bearish trend in the Bitcoin market.
